```python
import pandas as pd
import numpy as np
import heapq
netflix_df = pd.read_csv("netflixData.csv")
```

# Assignment: 
Show the most number of tv-shows and movies released by each country in descending order in Netflix dataset shared in class. Also plot a count plot of each country.

```python
dataset = netflix_df.groupby("Content Type")["Production Country"].value_counts()
```

```python
dataset
```

```python
country = list(dataset.Movie.index.values)
country
```

```python
countryValues = list(dataset.Movie.values)
countryValues
```

# Seprating the top values to be plotted as we have huge number of countries so, we are going to plot top 15 countries, values can be change as per need

```python
countryplt = heapq.nlargest(15,dataset)
countryplt
```

```python
countrypltName = country[:15]
countrypltName
```

# Plotting

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(15,8))
plt.plot(countrypltName,countryplt, '-ok', color='black')
```

```python
import seaborn as sns
```

```python
sns.countplot(x='Production Country',data=netflix_df,order=pd.value_counts(netflix_df['Production Country']).iloc[:15].index)
```
